```python
# see in the data set we have some columns which contain string as value
# to apply regression(Logistic) we need to convert this string
#value as discret/categorical value because logistic reg. just take 2 values(0/1)
```

```python
sna_data.head(2)
```

```python
# use pandas over sex coloumns to combind male and female into one columns
pd.get_dummies(sna_data['Gender'])
```

```python
# use pandas over sex coloumns to combind male and female into one columns
sex=pd.get_dummies(sna_data['Gender'], drop_first=True)
```

```python
sex.head(5)
```

```python
sna_data=pd.concat([sna_data,sex],axis=1)
sna_data.head(5)
```

```python
#Drop unnecessary columns like pclass, embarked, sex
sna_data.drop(['Gender'], axis=1, inplace=True)
```

```python
sna_data.head(2)
```

###Train Data

```python
X=sna_data.drop('Purchased', axis=1)
y=sna_data['Purchased']
```

```python
X
```

```python
#training the logistic model regarding the dataset pattern such that it can predict in future
from sklearn.model_selection import train_test_split
```

```python
y
```

```python
#dividing the dataset into test and train sets in a 3:7 ratio
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
x1=X_test.head(5)
```

```python
y1=y_test.head(5)
```

```python
#training the dataset values
logmodel.fit(X_train, y_train)
```

```python
#predicting y_pred based on x_test
predictions=logmodel.predict(X_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import classification_report
```

```python
y_test
```

```python
#building a classification report to determine our predictions accuracy
classification_report(y_test, predictions)
```

```python
# to determine accuracy we can use confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
#calculation of accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predictions)
```
